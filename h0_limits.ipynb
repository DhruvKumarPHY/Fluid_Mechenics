```python
import sympy as smp
from sympy import *
```

```python
gamma = smp.Symbol("\delta \gamma")
gamma
```

```python
c, A, P = smp.symbols('c A P')
```

# Alpha

```python
alpha = (((1-(pi**2)/(3*A)))) * ((pi**2) * (gamma*c)/2 ) - (pi**2 * (1/P)+(pi**2/(3*A)))
alpha
```

```python
# h ---> 0 
s1 = smp.limit(alpha,P,oo)
s1
```

```python
s2 = smp.limit(s1,A,0)
s2
```

```python
s1p =  smp.limit(alpha,A,0)
s1p
```

```python
s2p = smp.limit(s1p,P,oo)
s2p
```

```python
# h ---> oo
s3 = smp.limit(alpha,P,0)
s3
```

```python
s3p = smp.limit(alpha,A,oo)
s3p
```

```python
s4p = smp.limit(s3p,P,0)
s4p
```

```python
s4 = smp.limit(s3,A,oo)
s4
```

# Beta

```python
beta = ((pi**6 / (3*A)) * (4/P - gamma*c/2)) * (1+ gamma * c/2)
beta
```

```python
# h ---> 0 
s1 = smp.limit(beta,P,oo)
s1
```

```python
s2 = smp.limit(s1,A,0)
s2
```

```python
s1p =  smp.limit(beta,A,0)
s1p
```

```python
s2p = smp.limit(s1p,P,oo)
s2p
```

```python
# h ---> oo
s3 = smp.limit(beta,P,0)
s3
```

```python
s4 = smp.limit(s3,A,oo)
s4
```

```python
s3p = smp.limit(beta,A,oo)
s3p
```

```python
s4p = smp.limit(s3p,P,0)
s4p
```
