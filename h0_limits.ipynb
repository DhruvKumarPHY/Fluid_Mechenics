```python
import sympy as smp
from sympy import *
```

```python
gamma_d = smp.Symbol("\delta \gamma")
gamma_d
```

```python
c0, A, Pe = smp.symbols('c0 A P_{e}')
```

# Alpha

```python
alpha = (((1-(pi**2)/(3*A)))) * ((pi**2) * (gamma_d*c0)/2 ) - (pi**2 *(1/Pe)+(pi**4/(3*A)))
alpha
```

```python
# h ---> 0 
s1 = smp.limit(alpha,Pe,oo)
s1
```

```python
s2 = smp.limit(s1,A,0)
s2
```

```python
s1p =  smp.limit(alpha,A,0)
s1p
```

```python
s2p = smp.limit(s1p,Pe,oo)
s2p
```

```python
# h ---> oo
s3 = smp.limit(alpha,Pe,0)
s3
```

```python
s3p = smp.limit(alpha,A,oo)
s3p
```

```python
s4p = smp.limit(s3p,Pe,0)
s4p
```

```python
s4 = smp.limit(s3,A,oo)
s4
```

# n_max

```python
shi_1 = (alpha + ((alpha**2 - beta)**smp.Rational(1/2)))/2
shi_1
```

```python
shi_2 =  (alpha - ((alpha**2 - beta)**smp.Rational(1/2)))/2
shi_2
```

```python
n_max = ((pi**2 * (gamma_d*c0))/(2*(shi_1 - shi_2))) * ((smp.exp((shi_1/(shi_1 - shi_2))* smp.ln(shi_2/shi_1)) - (smp.exp((shi_2/(shi_1 - shi_2))* smp.ln(shi_2/shi_1)))))
n_max
```
